```python
from dotenv import load_dotenv
import os

load_dotenv(override=True)

print("Prospects DB:", os.getenv("PROSPECTS_DB_ID"))
print("Activity DB:", os.getenv("ACTIVITY_DB_ID"))
```

```python
from notion_client import Client

notion = Client(auth=os.getenv("NOTION_TOKEN"))

print("Connected to Notion client!")
```

```python
def query_all_pages(notion, data_source_id):
    all_results = []
    start_cursor = None

    while True:
        kwargs = {"data_source_id": data_source_id}
        if start_cursor:
            kwargs["start_cursor"] = start_cursor

        response = notion.data_sources.query(**kwargs)
        all_results.extend(response["results"])

        if not response.get("has_more"):
            break

        start_cursor = response.get("next_cursor")

    return all_results
```

```python
prospect_pages = query_all_pages(notion, os.getenv("PROSPECTS_DB_ID"))
activity_pages = query_all_pages(notion, os.getenv("ACTIVITY_DB_ID"))

print("Prospects:", len(prospect_pages))
print("Activity:", len(activity_pages))
```
